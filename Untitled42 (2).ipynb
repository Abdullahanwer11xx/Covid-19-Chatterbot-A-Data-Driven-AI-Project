***Step 1: Import Libraries***

```python
import json
import random
import pickle
import tkinter as tk
from tkinter import scrolledtext
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
```

***Step 2: Load the Intents Data***

```python
# Load the intents data from the JSON file
def load_intents(file_path):
    with open(file_path, 'r') as file:
        return json.load(file)
```

***Step 3: Prepare the Dataset***

```python
# Prepare the dataset: Convert intents into training data
def prepare_dataset(intents_data):
    texts = []
    labels = []
    for intent in intents_data['intents']:
        for pattern in intent['text']:
            texts.append(pattern)
            labels.append(intent['intent'])
    return texts, labels
```

***Step 4: Train the Model***

```python
# Train the model using Naive Bayes classifier
def train_model(texts, labels):
    vectorizer = CountVectorizer()
    X = vectorizer.fit_transform(texts)
    X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.2, random_state=42)

    model = MultinomialNB()
    model.fit(X_train, y_train)

    # Evaluate the model
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Model trained with accuracy: {accuracy:.2f}")
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred))

    # Save the model and vectorizer
    with open('model.pkl', 'wb') as model_file:
        pickle.dump(model, model_file)

    with open('vectorizer.pkl', 'wb') as vectorizer_file:
        pickle.dump(vectorizer, vectorizer_file)

    # Plot confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred, labels=model.classes_)
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=model.classes_, yticklabels=model.classes_)
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    # Plot accuracy graph
    plt.figure(figsize=(10, 5))
    plt.plot([accuracy], marker='o', color='blue', label='Model Accuracy')
    plt.title("Model Accuracy Over Training")
    plt.ylabel("Accuracy")
    plt.xlabel("Epochs")
    plt.legend()
    plt.show()

    # Plot word frequency graph
    plot_word_frequency(texts)

def plot_word_frequency(texts):
    words = ' '.join(texts).split()
    word_freq = Counter(words)
    common_words = word_freq.most_common(20)
    words, frequencies = zip(*common_words)
    plt.figure(figsize=(10, 6))
    plt.bar(words, frequencies, color='skyblue')
    plt.xticks(rotation=90)
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.title('Top 20 Most Frequent Words in the Dataset')
    plt.show()
```

***Step 6: Predict Intent Based on User Input***

```python
# Function to predict the intent and return a response
def predict_intent(model, vectorizer, user_input):
    user_input_transformed = vectorizer.transform([user_input])
    prediction = model.predict(user_input_transformed)
    intent = prediction[0]
    return get_response(intent)
```

***Step 7: Build the GUI for the Chatbot***

```python
# Create the GUI for the chatbot
class ChatbotGUI:
    def __init__(self, root, model, vectorizer):
        self.root = root
        self.root.title("COVID-19 Medical Chatbot")
        self.model = model
        self.vectorizer = vectorizer
        self.chat_history = scrolledtext.ScrolledText(root, wrap=tk.WORD, state=tk.DISABLED, width=60, height=20)
        self.chat_history.grid(row=0, column=0, padx=10, pady=10)
        self.entry_box = tk.Entry(root, width=50)
        self.entry_box.grid(row=1, column=0, padx=10, pady=10)
        self.send_button = tk.Button(root, text="Send", width=20, command=self.send_message)
        self.send_button.grid(row=1, column=1, padx=10, pady=10)

    def send_message(self):
        user_input = self.entry_box.get()
        if user_input.lower() == 'exit':
            self.chat_history.config(state=tk.NORMAL)
            self.chat_history.insert(tk.END, "Goodbye!\n")
            self.chat_history.config(state=tk.DISABLED)
            self.root.destroy()  # Gracefully close the window
            return

        # Display user input in the chat history
        self.chat_history.config(state=tk.NORMAL)
        self.chat_history.insert(tk.END, f"You: {user_input}\n")
        self.chat_history.config(state=tk.DISABLED)

        # Get the bot's response
        response = predict_intent(self.model, self.vectorizer, user_input)

        # Display bot's response in the chat history
        self.chat_history.config(state=tk.NORMAL)
        self.chat_history.insert(tk.END, f"Bot: {response}\n")
        self.chat_history.config(state=tk.DISABLED)

        # Clear the entry box
        self.entry_box.delete(0, tk.END)
```

***Step 8: Main Function to Load Data, Train Model, and Start GUI***

```python
# Load the trained model and vectorizer
def load_model_and_vectorizer():
    with open('model.pkl', 'rb') as model_file:
        model = pickle.load(model_file)
    with open('vectorizer.pkl', 'rb') as vectorizer_file:
        vectorizer = pickle.load(vectorizer_file)
    return model, vectorizer
```
